### Prapare Windows

```python
dt.reset_index(inplace=True)
testdf.reset_index(inplace=True)
```

```python
def prepare_windows(data, windows_size=6):
    window_features = []
    for idx in data.index:
        left_fetures = [data.iloc[x].x if x >= 0 else 0 for x in range(idx-windows_size,idx)]
        right_fetures = [data.iloc[x].x if x <= data.shape[0]-1 else 0 for x in range(idx+1,idx+windows_size+1)]
        window_features.append(left_fetures + right_fetures)
    return pd.DataFrame(window_features)
```

```python
windows_features_df = prepare_windows(dt)
windows_features_test_df = prepare_windows(testdf)
```

```python
windows_features_df.head(3)
```

```python
def q1(x):
    return x.quantile(0.05)

def q2(x):
    return x.quantile(0.25)

def q3(x):
    return x.quantile(0.75)


def q4(x):
    return x.quantile(0.95)
```

```python
windows_metrics_df = windows_features_df.agg(
    ["mean","min",'max','sum','median', q1,q2,q3, q4]
    , axis = 1
)

windows_metrics_test_df = windows_features_test_df.agg(
    ["mean","min",'max','sum','median', q1,q2,q3, q4]
    , axis = 1
)
```

```python
windows_metrics_df.head(3)
```

```python
dt_fetures = pd.concat([dt,windows_features_df,windows_metrics_df],axis=1)

dt_fetures_test = pd.concat([testdf,windows_features_test_df,windows_metrics_test_df],axis=1)
dt_fetures_test.head(10)
```

```python
dt_fetures.head(10)
```

```python
cols_drop = ["index", "time","id","y","level_0"]
cols_drop_test = ["index", "time","id","level_0"]
```

```python
y = dt_fetures.y
X = dt_fetures.drop(cols_drop, axis=1)
X.head(3)
```

```python
X_test = dt_fetures_test.drop(cols_drop_test, axis=1)
X_test.head(3)
```

```python
X_train, X_val,y_train,y_val = train_test_split(X,y, test_size=0.15, random_state=42)

X_train.shape, X_val.shape
```

```python
model = CatBoostClassifier(verbose=0,
                           learning_rate=0.3,
                           l2_leaf_reg = 100,
                           depth=10,
                           task_type='GPU')
```

```python
model.fit(X_train, y_train)
```

```python
for x in [.1,.2,.25,.3,.35,.4,.5,.6,.7,.8]:
    y_pred = (model.predict(X_val) >= x).astype(int)
    f1_score_val = f1_score(y_val.to_numpy(), y_pred, average='micro')
    print(f'f1_score: {f1_score_val}')
    print(x, classification_report(y_val,y_pred))
```

### Torch model

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

from sklearn.preprocessing import StandardScaler    
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python
class TrainData(Dataset):
    
    def __init__(self, X_data, y_data):
        self.X_data = X_data
        self.y_data = y_data
        
    def __getitem__(self, index):
        return self.X_data[index], self.y_data[index]
        
    def __len__ (self):
        return len(self.X_data)
    
class TestData(Dataset):
    
    def __init__(self, X_data):
        self.X_data = X_data
        
    def __getitem__(self, index):
        return self.X_data[index]
        
    def __len__ (self):
        return len(self.X_data)
    
train_data = TrainData(torch.FloatTensor(X_train.to_numpy()), torch.FloatTensor(y_train.to_numpy()))
val_data = TestData(torch.FloatTensor(X_val.to_numpy())) 
test_data = TestData(torch.FloatTensor(X_test.to_numpy())) 
```

```python
X_test
```

```python
X_train
```

```python
EPOCHS = 200
BATCH_SIZE = 512
LEARNING_RATE = 0.001
```

```python
train_loader = DataLoader(dataset=train_data, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(dataset=test_data, batch_size=1)
val_loader = DataLoader(dataset=val_data, batch_size=1)
```

```python
class BinaryClassification(nn.Module):
    def __init__(self, input_features=12):
        super(BinaryClassification, self).__init__()
        # Number of input features is 12.
        self.layer_1 = nn.Linear(input_features, 128) 
        self.layer_2 = nn.Linear(128, 256)
        self.layer_3 = nn.Linear(256, 256)
        self.layer_4 = nn.Linear(256, 512)
        self.layer_5 = nn.Linear(512, 512)
        self.layer_6 = nn.Linear(512, 128)
        self.layer_7 = nn.Linear(128, 64)
        self.layer_out = nn.Linear(64, 1) 
        
        self.relu = nn.ReLU()
        self.dropout1 = nn.Dropout(p=0.3)
        self.dropout2 = nn.Dropout(p=0.3)
        self.dropout3 = nn.Dropout(p=0.2)
        self.dropout4 = nn.Dropout(p=0.2)    
        self.dropout5 = nn.Dropout(p=0.2)   
        self.dropout6 = nn.Dropout(p=0.2)         
        self.batchnorm1 = nn.BatchNorm1d(128)
        self.batchnorm2 = nn.BatchNorm1d(256)
        self.batchnorm3 = nn.BatchNorm1d(256)
        self.batchnorm4 = nn.BatchNorm1d(512)
        self.batchnorm5 = nn.BatchNorm1d(512)
        self.batchnorm6 = nn.BatchNorm1d(128)
        self.batchnorm7 = nn.BatchNorm1d(64)
        
    def forward(self, inputs):
        x = self.relu(self.layer_1(inputs))
        x = self.batchnorm1(x)
        x = self.relu(self.layer_2(x))
        x = self.batchnorm2(x)        
        x = self.dropout1(x)
        x = self.relu(self.layer_3(x))
        x = self.batchnorm3(x)
        x = self.dropout2(x)
        x = self.relu(self.layer_4(x))
        x = self.batchnorm4(x)
        x = self.dropout3(x)
        x = self.relu(self.layer_5(x))
        x = self.batchnorm5(x)
        x = self.dropout4(x)
        x = self.relu(self.layer_6(x))
        x = self.batchnorm6(x)
        x = self.dropout5(x)
        x = self.relu(self.layer_7(x))
        x = self.batchnorm7(x)
        x = self.dropout6(x)
        
        x = self.layer_out(x)
        
        return x
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
model = BinaryClassification(input_features = X.shape[1])
model.to(device)
print(model)
```

```python
def binary_acc(y_pred, y_test):
    y_pred_tag = torch.round(torch.sigmoid(y_pred))

    correct_results_sum = (y_pred_tag == y_test).sum().float()
    acc = correct_results_sum/y_test.shape[0]
    acc = torch.round(acc * 100)
    
    return acc
```

```python
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE)
scheduler = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts(optimizer,T_0=len(train_loader),T_mult=2)
```

```python
def validate(model,test_loader):
    y_pred_list = []
    model.eval()
    with torch.no_grad():
        for X_batch in test_loader:
            X_batch = X_batch.to(device)
            y_test_pred = model(X_batch)
            y_test_pred = torch.sigmoid(y_test_pred)
            y_pred_tag = (y_test_pred>0.3).int()
            y_pred_list.append(y_pred_tag.cpu().numpy())

    y_pred_list = [a.squeeze().tolist() for a in y_pred_list]
    return f1_score(y_test, y_pred_list, average='micro')
```

```python
best_f1_score  = 0
for e in range(1, EPOCHS+1):
    model.train()
    epoch_loss = 0
    epoch_acc = 0
    epoch_f1_score = 0
    f1_score_val = 0
    iters = len(train_loader)
    i = 0
    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        optimizer.zero_grad()
        y_pred = model(X_batch)
        loss = criterion(y_pred, y_batch.unsqueeze(1))
        acc = binary_acc(y_pred, y_batch.unsqueeze(1))
        
        y_pred_tag = (torch.sigmoid(y_pred)>0.3).int()
        f1 = f1_score(y_pred_tag.detach().cpu(), y_batch.unsqueeze(1).detach().cpu().numpy(), average='micro')
        
        loss.backward()
        optimizer.step()
        scheduler.step(e + i / iters)
        i += 1
        epoch_loss += loss.item()
        epoch_acc += acc.item()
        epoch_f1_score += f1
        current_f1_dcore = epoch_f1_score/len(train_loader)
        
    if e%10==0:
        print(f'Epoch {e+0:03}: | Loss: {epoch_loss/len(train_loader):.5f} | Acc: {epoch_acc/len(train_loader):.3f} | f1: {epoch_f1_score/len(train_loader):.3f} | f1 val: {f1_score_val} ')
```

```python
y_pred_list = []
model.eval()
with torch.no_grad():
    for X_batch in val_loader:
        X_batch = X_batch.to(device)
        y_test_pred = model(X_batch)
        y_test_pred = torch.sigmoid(y_test_pred)
        y_pred_tag = (y_test_pred>0.6).int()
        y_pred_list.append(y_pred_tag.cpu().numpy())

y_pred_list = [a.squeeze().tolist() for a in y_pred_list]
```
